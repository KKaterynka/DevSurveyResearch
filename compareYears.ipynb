##**Load the data from different years**

```python
df_survey_2022 = pd.read_csv("/content/drive/MyDrive/survey_results_public2022.csv", sep=',')
df_survey_2021 = pd.read_csv("/content/drive/MyDrive/survey_results2021.csv", sep=',')
df_survey_2020 = pd.read_csv("/content/drive/MyDrive/survey_results_public2020.csv", sep=',')
df_survey_2019 = pd.read_csv("/content/drive/MyDrive/survey_results_public2019.csv", sep=',')
df_survey_2018 = pd.read_csv("/content/drive/MyDrive/survey_results_public2018.csv", sep=',')
```

```python
df_survey_2022_salary = df_survey_2022.ConvertedCompYearly
df_survey_2021_salary = df_survey_2021.ConvertedCompYearly
df_survey_2020_salary = df_survey_2020.ConvertedComp
df_survey_2019_salary = df_survey_2019.ConvertedComp
df_survey_2018_salary = df_survey_2018.ConvertedSalary
```

##**Find the median value of salary in each year**

```python
mean_salary_2022 = df_survey_2022_salary.dropna().median()
mean_salary_2021 = df_survey_2021_salary.dropna().median()
mean_salary_2020 = df_survey_2020_salary.dropna().median()
mean_salary_2019 = df_survey_2019_salary.dropna().median()
mean_salary_2018 = df_survey_2018_salary.dropna().median()
```

##**Visualize and analyse**

```python
# define the years and mean salaries
years = [2018, 2019, 2020, 2021, 2022]
mean_salaries = [mean_salary_2018, mean_salary_2019, mean_salary_2020, mean_salary_2021, mean_salary_2022]

# create the line plot
plt.plot(years, mean_salaries, marker='o', linestyle='-', color='#2CBDFE')
sns.set('notebook', palette='colorblind')

# set plot title and labels
plt.title('Median Salary by Year')
plt.xlabel('Year')
plt.ylabel('Median Salary')

# show the plot
plt.xticks(years)
plt.show()
```

From this graph we can conclude, that the median annual salary of the respondents of 2018-2021 years was quite stable on the value of \~56000. But the results of the research in 2022, showed, that the median annual salary value has grown to \~68000.

One of the reasons, of such change can be the increasing popularity of development specialists, so to get the professional companies are increasing the amount of salary.

The other reason, can be that the respondents of this research changed. Either the amount and the variety of highly paid developers increased, or the previous respondents got promotions and pay raises.
